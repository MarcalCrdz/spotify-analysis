```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df["track_genre"].value_counts().head(10)
```

```python
top_generos = df["track_genre"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_generos.values,
    y=top_generos.index
)

plt.title("Top 10 gêneros mais frequentes")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="energy",
    y="popularity"
)

plt.title("Energia vs Popularidade")
plt.show()
```

```python
sns.histplot(df["popularity"], bins=30)

plt.title("Distribuição da popularidade")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap="coolwarm"
)

plt.show()
```
